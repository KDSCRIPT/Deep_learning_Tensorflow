```python
#@title Default title text
#create a rank 4 tensor
rank_4_tensor=tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape,rank_4_tensor.ndim,tf.size(rank_4_tensor)
```

```python
print("Datatype of each element:",rank_4_tensor.dtype)
print("Number of dimensions:",rank_4_tensor.ndim)
print("Shape of tensor:",rank_4_tensor.shape)
print("Elements along the 0 axis:",rank_4_tensor[0])
print("Elements along last axis:",rank_4_tensor[-1])
print("Total number of elements in our tensor:",tf.size(rank_4_tensor).numpy())
```

```python
#getting first 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
#get first element each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1]
```

```python
#create a rank 2 tensor
rank_2_tensor=tf.constant([[10,7],[3,4]])
rank_2_tensor.shape,rank_2_tensor.ndim
```

```python
#get last item of each row of our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
#add dimensions to our rank 2 tensors
rank_3_tensor=rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
#alternative for tf.newaxis
tf.expand_dims(rank_2_tensor,axis=-1,)
```
